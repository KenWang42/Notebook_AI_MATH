```python
def KWARG(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} == {value}")
```

```python
KWARG(first = 'F', second = 'U', third = 'C', fourth = 'K')
```

```python
def Four(arg1, arg2, arg3, arg4):
    print('arg1 is ', arg1)
    print('arg2 is ', arg2)
    print('arg3 is ', arg3)
    print('arg1 is ', arg4)
```

```python
args = ('F', 'U', 'C', 'K')
Four(*args)
```

```python
kwargs = {'arg1':'F', 'arg2':'U', 'arg3':'C', 'arg4':'K'}
Four(**kwargs)
```
