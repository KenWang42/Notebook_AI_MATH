### groupby

```python
ufo = pd.read_csv("http://bit.ly/uforeports")
ufo.head()
```

```python
ufo_state = ufo.groupby('State')
```

```python
ufo_state.count()
```

```python
ufo_state = ufo_state.count()
```

```python
ufo_state.sort_values(by = "Time",
                     ascending = False,
                     inplace = True)
ufo_state[:20].Time.plot(kind = "bar")
```

```python
ufo.groupby(["State", "City"]).count()
```

```python
ufo.groupby(["State", "City"]).count().reset_index()
```

### datetime

```python
ufo['datetime'] = pd.to_datetime(ufo.Time)
```

```python
ufo.head()
```

```python
import datetime as dt
```

```python
ufo["month"] = ufo.datetime.dt.month
```

```python
ufo.head()
```

```python
ufo["weekday"] = ufo.datetime.dt.weekday_name
```

```python
ufo.head()
```

```python
ufo_weekdays = ufo.groupby("weekday").count()
```

```python
ufo_weekdays.sort_values(by = "Time",
                        ascending = False,
                        inplace = True)
```

```python
ufo_weekdays.Time.plot(kind = "bar")
```

```python
ufo_months = ufo.groupby("month").count()
```

```python
ufo_months.sort_values(by = "Time",
                        ascending = False,
                        inplace = True)
```

```python
ufo_months.Time.plot(kind = "bar")
```
